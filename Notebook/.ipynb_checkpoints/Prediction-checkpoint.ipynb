```python
import pickle 
import joblib
import numpy as np
import base64
import cv2
import pywt
import joblib
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python

global __model
global __class_number_to_name
with open('./ML logs/saved_model.pkl', 'rb') as f:
     __model = joblib.load(f)
```

```python
__class_number_to_name = {1:'Male',0:'Female'}
```

```python

def w2d(img, mode='haar', level=1):
    imArray = img

    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255.0;

    coeffs=pywt.wavedec2(imArray, mode, level=level)


    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255.0;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python

def pred(path):
    img = cv2.imread(path)
    plt.imshow(img)
    scalled_raw_img = cv2.resize(img, (32, 32))
    img_har = w2d(img,'db1',5)
    scalled_img_har = cv2.resize(img_har, (32, 32))
    test_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
    return __class_number_to_name[__model.predict(test_img.reshape(1,-1))[0]]
```

```python
pred('./test.png')
```

```python
val_lbl_npy = np.load('../Data/val_labels.npy')
val_img_npy = np.load('../Data/val_images.npy')
subdir = ['Male','Female']
path = '../Data/Gender detection/Test - Copy/'
X_test, y_test, ids = [], [], []
for gender in subdir:
    for img_name in os.listdir(path+gender):
        img = cv2.imread(path+gender+'/'+img_name)
        scalled_raw_img = cv2.resize(img, (32, 32))
        img_har = w2d(img,'db1',5)
        scalled_img_har = cv2.resize(img_har, (32, 32))
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        X_test.append(combined_img)
        y_test.append(int((val_lbl_npy[int(img_name.split('.')[0])]).split('-')[0]))
        ids.append(int(img_name.split('.')[0]))
        
```

```python
X_test = np.array(X_test).reshape(len(X_test),4096).astype(float)
y_test = np.array(y_test).astype(int)
ids = np.array(ids).astype(int)
```

```python
X_test.shape, y_test.shape, ids.shape
```

```python
y_pred = __model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns 
from sklearn.metrics import classification_report

cm = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(8,8))
g = sns.heatmap(cm,annot=True,ax=ax, fmt='0.1f',cmap='Accent_r')
g.set_yticklabels(g.get_yticklabels(), rotation = 0, fontsize = 12)
g.set_xticklabels(g.get_xticklabels(), rotation = 90, fontsize = 12)
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
```

```python
print(classification_report(y_test, y_pred))
```

```python
df = pd.DataFrame([ids, y_pred])
df = df.T
df.columns = ['ID', 'Prediction']
df.sort_values(by='ID',ascending=True, inplace=True)
df.head()
```

```python
df.to_csv('./ML logs/public_test_submission.csv',index=False)
```
