```python

for i in range(train_lbl.shape[0]):
    
    data = im.fromarray(train_img[i])
    if train_lbl[i].split('-')[0]=='1':
        data.save('../Data/Gender detection/Train/Male/{}.png'.format(i))
    else:
        data.save('../Data/Gender detection/Train/Female/{}.png'.format(i))

```

```python

for i in range(val_lbl.shape[0]):
    
    data = im.fromarray(val_img[i])
    if val_lbl[i].split('-')[0]=='1':
        data.save('../Data/Gender detection/Test/Male/{}.png'.format(i))
    else:
        data.save('../Data/Gender detection/Test/Female/{}.png'.format(i))

```

# Age recognition data population

```python

for i in range(train_lbl.shape[0]):
    
    data = im.fromarray(train_img[i])
    thershold = int(train_lbl[i].split('-')[1])
    if (thershold>=6) and (thershold<=20):
        data.save('../Data/Age Recognition/Train/6-20/{}.png'.format(i))
    elif (thershold>=25) and (thershold<=30):
        data.save('../Data/Age Recognition/Train/25-30/{}.png'.format(i))
    elif (thershold>=42) and (thershold<=48):
        data.save('../Data/Age Recognition/Train/42-48/{}.png'.format(i))
    else:
        data.save('../Data/Age Recognition/Train/60-98/{}.png'.format(i))
```

```python

for i in range(val_lbl.shape[0]):
    
    data = im.fromarray(val_img[i])
    thershold = int(val_lbl[i].split('-')[1])
    if (thershold>=6) and (thershold<=20):
        data.save('../Data/Age Recognition/Test/6-20/{}.png'.format(i))
    elif (thershold>=25) and (thershold<=30):
        data.save('../Data/Age Recognition/Test/25-30/{}.png'.format(i))
    elif (thershold>=42) and (thershold<=48):
        data.save('../Data/Age Recognition/Test/42-48/{}.png'.format(i))
    else:
        data.save('../Data/Age Recognition/Test/60-98/{}.png'.format(i))
```
